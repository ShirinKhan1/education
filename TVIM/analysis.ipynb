```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

https://www.kaggle.com/datasets/hasibur013/diabetes-dataset

Столбцы:
* Pregnancies  (целое число): Количество раз, когда пациентка была беременна.\n
* Glucose  (целое число): концентрация глюкозы в плазме крови (мг/дл) после 2-часового перорального теста на толерантность к глюкозе.
* BloodPressure (целое число): диастолическое артериальное давление (мм рт.ст.).
* SkinThickness  (целое число): Толщина кожной складки на трехглавой мышце (мм).
* Insulin  (целое число): суточная доза сывороточного инсулина (мкме/мл).
* BMI (плавающее значение): Индекс массы тела, определяемый как вес в кг/(рост в метрах).^2.
* DiabetesPedigreeFunction  (с плавающей точкой): показатель, указывающий на генетическую предрасположенность к диабету, основанный на семейном анамнезе.
* Age (целое число): Возраст пациента (в годах).
* Outcome (двоичный): Целевая переменная, где 1 указывает на диабет, а 0 - на отсутствие диабета.

```python
df = pd.read_csv('diabetes_dataset.csv')
df.head()
```

```python
skinthickness = df['SkinThickness'].dropna()

plt.hist(skinthickness, bins=15, density=True, alpha=0.6, color="b", label="Толщина кожной складки на трехглавой мышце")

mu_2, std_2 = skinthickness.mean(), skinthickness.std()

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 65)
p = norm.pdf(x, mu_2, std_2)

plt.plot(x, p, "k", linewidth=2, label="Нормальное распределение")

plt.title("Распределение кожной складки (толщина)")
plt.xlabel("Толщина кожной сладки")
plt.ylabel("Плотность")
plt.legend()
plt.show()
```

```python
mean_value_2 = df['SkinThickness'].mean()
print(f"Математическое ожидание: {mean_value_2:.3f}")

variance_2 = df['SkinThickness'].var()
print(f"Дисперсия: {variance_2:.3f}")

skewness_2 = df['SkinThickness'].skew()
print(f"Асимметрия: {skewness_2:.3f}")

kurtosis_2 = df['SkinThickness'].kurtosis()
print(f"Эксцесс: {kurtosis_2:.3f}")

quantile_05_2 = df['SkinThickness'].quantile(0.05)
quantile_95_2 = df['SkinThickness'].quantile(0.95)
print(f"Квантиль 0,05: {quantile_05_2:.3f}")
print(f"Квантиль 0,95: {quantile_95_2:.3f}")

quantile_025_2 = df['SkinThickness'].quantile(0.025)
print(f"2,5%-ная точка: {quantile_025_2:.6f}")
```
